```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

# Aseg√∫rate de tener descargadas las stopwords de nltk
nltk.download('stopwords')

stopwords_libros = {'book', 'read', 'novel', 'story', 'books', 'reading', 'author', 'chapter', 'reader', 'reading', 'writting', 'one', 'quot', 'would', 'review'}

# Obtener las stopwords comunes en ingl√©s de NLTK
stopwords_comunes = set(stopwords.words('english'))

stopwords_verbos= {'like', 'time', 'know', 'make', 'people', 'good', 'really', 'think', 'get', 'want', 'love', 'seem', 'much', 'even', 'work', 'take', 'find', 'first', 'lot', 'also', 'see', 'really', 'character'}

# Combinar ambas listas de stopwords
stopwords_totales = stopwords_libros.union(stopwords_comunes)
stopwords_totales = stopwords_totales.union(stopwords_verbos)

# Asumimos que 'categoria' es la columna con las categor√≠as (Positivo, Negativo, Neutral)
# Filtramos las rese√±as positivas
positivas = df[df['categoria'] == 'positivo']['review/text'].str.cat(sep=" ")

# Configuramos la nube de palabras, incluyendo las stopwords combinadas
wc = WordCloud(
    width=500,
    height=500,
    min_font_size=15,
    background_color='white',
    stopwords=stopwords_totales  # A√±adimos las stopwords combinadas
)

# Nube de palabras para las rese√±as positivas
positivas_wc = wc.generate(positivas)
# Mostramos la nube de palabras positiva
plt.figure(figsize=(7, 7))
plt.axis('off')
plt.title('Nube de palabras - Positivas')
plt.imshow(positivas_wc)
plt.show()

```

```python
import re
import html

# Lista de stopwords personalizadas
stopwords_libros = {'book', 'read', 'novel', 'story', 'books', 'reading', 'author', 'chapter', 'reader', 'reading', 'writing', 'one', 'quot', 'would', 'review', 'page', 'word'}

# Obtener las stopwords comunes en ingl√©s de NLTK
from nltk.corpus import stopwords
stopwords_comunes = set(stopwords.words('english'))

# Lista de stopwords de verbos
stopwords_verbos = {'time', 'know', 'make', 'people', 'good', 'really', 'think', 'get', 'want', 'seem', 'much', 'even', 'work', 'take', 'find', 'first', 'lot', 'also', 'see', 'really'}

# Combinar todas las listas de stopwords
stopwords_totales = stopwords_libros.union(stopwords_comunes).union(stopwords_verbos)

# Convertir el conjunto de stopwords a lista
stopwords_totales = list(stopwords_totales)


```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

## Tokenizaci√≥n y an√°lisis de palabras

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# Preprocesamiento con TfidfVectorizer usando las stopwords personalizadas
vectorizer = TfidfVectorizer(stop_words=stopwords_totales)  # Pasamos las stopwords personalizadas

# Transformar los datos
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
import numpy as np


# Entrenar el modelo de regresi√≥n log√≠stica con los datos transformados (y_train_onehot)
modelo_lr = LogisticRegression(max_iter=1000)
modelo_lr.fit(X_train_vec, y_train)  # X_train_vec son las caracter√≠sticas, y_train_onehot son las etiquetas codificadas

# Realizar predicciones sobre el conjunto de prueba
y_pred = modelo_lr.predict(X_test_vec)

# Evaluar el rendimiento del modelo (usando precisi√≥n en este caso)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
# Mostrar la matriz de confusi√≥n
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=modelo_lr.classes_, yticklabels=modelo_lr.classes_)
plt.title("Matriz de Confusi√≥n")
plt.xlabel("Predicciones")
plt.ylabel("Verdaderas Etiquetas")
plt.show()

# Reporte de clasificaci√≥n: precisi√≥n, recuperaci√≥n y F1 para cada clase
print("\nReporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Definir los par√°metros que queremos probar
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Valores de C (regularizaci√≥n)
    'solver': ['lbfgs', 'liblinear', 'saga'],  # Solvers disponibles
    'penalty': ['l2'],  # Tipo de regularizaci√≥n (l2 o l1)
    'max_iter': [100, 200],  # N√∫mero m√°ximo de iteraciones
}

# Crear el modelo de regresi√≥n log√≠stica
modelo_lr = LogisticRegression()

# Crear el objeto GridSearchCV con validaci√≥n cruzada de 5 pliegues
grid_search = GridSearchCV(estimator=modelo_lr, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# Ajustar el GridSearch con los datos de entrenamiento
grid_search.fit(X_train_vec, y_train)

# Ver los mejores par√°metros encontrados
print("Mejores par√°metros:", grid_search.best_params_)

# Evaluar el rendimiento del modelo ajustado
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_vec)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy despu√©s del ajuste de hiperpar√°metros: {accuracy:.4f}")
```
